1. Ingresos Totales de las Tiendas
El análisis de los ingresos de cada tienda muestra la cantidad total generada por ventas en cada una. A continuación, se presentan los resultados de los ingresos totales:

Tienda 1: Ingresos totales de $1150880400.0

Tienda 2: Ingresos totales de $1116343500.0

Tienda 3: Ingresos totales de $1098019600.0

Tienda 4: Ingresos totales de $1038375700.0



```python
#Grafica de comparación de ganancias Anuales para las tiendas

# Convertimos las fechas a tipo datetime y agregamos el año
for df in [tienda, tienda2, tienda3, tienda4]:
    df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True)
    df['Año'] = df['Fecha de Compra'].dt.year

# Agrupamos y sumamos por año
ganancias_t1 = tienda.groupby('Año')['Precio'].sum()
ganancias_t2 = tienda2.groupby('Año')['Precio'].sum()
ganancias_t3 = tienda3.groupby('Año')['Precio'].sum()
ganancias_t4 = tienda4.groupby('Año')['Precio'].sum()

# Unimos todos en un solo DataFrame
df_ganancias = pd.DataFrame({
    'Tienda 1': ganancias_t1,
    'Tienda 2': ganancias_t2,
    'Tienda 3': ganancias_t3,
    'Tienda 4': ganancias_t4
}).fillna(0)

# Hacemos el gráfico
years = df_ganancias.index
bar_width = 0.2
x = range(len(years))

plt.figure(figsize=(12,6))

# Las 4 barras, desplazadas horizontalmente
plt.bar([i - 1.5*bar_width for i in x], df_ganancias['Tienda 1'], width=bar_width, label='Tienda 1')
plt.bar([i - 0.5*bar_width for i in x], df_ganancias['Tienda 2'], width=bar_width, label='Tienda 2')
plt.bar([i + 0.5*bar_width for i in x], df_ganancias['Tienda 3'], width=bar_width, label='Tienda 3')
plt.bar([i + 1.5*bar_width for i in x], df_ganancias['Tienda 4'], width=bar_width, label='Tienda 4')

# Etiquetas y estética
plt.xticks(x, years)
plt.xlabel('Año')
plt.ylabel('Ganancias Totales')
plt.title('Comparación de Ganancias Anuales por Tienda')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

A partir de estos ingresos, se puede observar que Tienda 1 lidera en términos de ventas totales, con un ingreso superior a las otras tiendas, lo que indica un alto rendimiento en términos de volumen de ventas.

2. Categorías de productos más y menos vendidas
En este análisis, se compararon las categorías de productos más y menos vendidas en las cuatro tiendas. Para hacer esto, se utilizó un gráfico de barras que muestra la frecuencia de ventas de cada categoría, con barras separadas para cada tienda.

```python
#Grafico de comparación de ventas por categoria de cada tienda

ventas_t1 = tienda['Categoría del Producto'].value_counts()
ventas_t2 = tienda2['Categoría del Producto'].value_counts()
ventas_t3 = tienda3['Categoría del Producto'].value_counts()
ventas_t4 = tienda4['Categoría del Producto'].value_counts()

df_ventas = pd.DataFrame({
    'Tienda 1': ventas_t1,
    'Tienda 2': ventas_t2,
    'Tienda 3': ventas_t3,
    'Tienda 4': ventas_t4
})

df_ventas = df_ventas.astype(int)

categorias = df_ventas.index
x = np.arange(len(categorias))

width = 0.2

plt.bar(x - 1.5*width, df_ventas['Tienda 1'], width=width, label='Tienda 1')
plt.bar(x - 0.5*width, df_ventas['Tienda 2'], width=width, label='Tienda 2')
plt.bar(x + 0.5*width, df_ventas['Tienda 3'], width=width, label='Tienda 3')
plt.bar(x + 1.5*width, df_ventas['Tienda 4'], width=width, label='Tienda 4')

plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.title('Ventas por Categoría en las 4 Tiendas')
plt.xticks(x, categorias, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

```

**Análisis:**

**Categoría más vendida:** Después de revisar todas las categorías de productos en cada tienda, Muebles se destacó como la categoría más vendida en las cuatro tiendas, con un volumen de ventas significativamente mayor que otras categorías. Esta categoría es la más popular, lo que indica que es un producto clave para las tiendas.

**Categoría menos vendida:** En cuanto a las categorías menos vendidas, Artículos para el hogar mostró un bajo rendimiento en general, aunque no en todas las tiendas. A pesar de que en la Tienda 4 esta categoría tuvo un rendimiento mejor que en las otras tiendas, al hacer la sumatoria total de ventas de la categoría en todas las tiendas, resultó ser la categoría con menor volumen de ventas.

3. Calificaciones promedio de los clientes por tienda

La calificación promedio otorgada por los clientes después de realizar una compra es un indicador fundamental para evaluar la calidad del servicio, la satisfacción del usuario y la experiencia general de compra.
